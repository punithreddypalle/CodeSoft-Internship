# <font color="Blue"> Handling the data

```python
#droping the column Time
cfd=cfd.drop(columns="Time",axis=1)
```

```python
cfd.columns
```

```python
x=cfd[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11',
       'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21',
       'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
```

```python
x.head()
```

```python
y=cfd["Class"]
```

```python
y.head()
```

# <font color="blue"> Splitting data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

# <font color="blue"> Models Training

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
y_pred_log=log_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
```

```python
log_acc=accuracy_score(y_test,y_pred_log)
log_acc
```

```python
print(confusion_matrix(y_test,y_pred_log))
```

```python
print(classification_report(y_test,y_pred_log))
```

```python
model_result={}
```

```python
model_result["Logistic Regression Default"]=accuracy_score(y_test,y_pred_log)
```

```python
model_result
```

## Logistic Regression with l1 

```python
log_reg_l1=LogisticRegression(penalty="l1",solver="saga")
log_reg_l1.fit(x_train,y_train)
```

```python
y_pred_log_l1=log_reg_l1.predict(x_test)
```

```python
log_l1_acc=accuracy_score(y_test,y_pred_log_l1)
log_l1_acc
```

```python
print(confusion_matrix(y_test,y_pred_log_l1))
```

```python
print(classification_report(y_test,y_pred_log_l1))
```

```python
model_result["Logistic Regression with l1"]=accuracy_score(y_test,y_pred_log_l1)
```

```python
model_result
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dct=DecisionTreeClassifier()
dct.fit(x_train,y_train)
```

```python
y_pred_dct=dct.predict(x_test)
```

```python
dct_acc=accuracy_score(y_test,y_pred_dct)
dct_acc
```

```python
print(confusion_matrix(y_test,y_pred_dct))
```

```python
print(classification_report(y_test,y_pred_dct))
```

```python
model_result["Decision Tree"]=accuracy_score(y_test,y_pred_dct)
```

```python
model_result
```

## K-Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_pred_knn=knn.predict(x_test)
```

```python
knn_acc=accuracy_score(y_test,y_pred_knn)
knn_acc
```

```python
print(confusion_matrix(y_test,y_pred_knn))
```

```python
print(classification_report(y_test,y_pred_knn))
```

```python
model_result["K-Neighbors"]=accuracy_score(y_test,y_pred_knn)
```

```python
model_result
```

## Bagging Classifier with Decision Tree

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag=BaggingClassifier(DecisionTreeClassifier())   
bag.fit(x_train,y_train)                          
```

```python
y_pred_bag=bag.predict(x_test)
```

```python
bag_acc=accuracy_score(y_test,y_pred_bag)
bag_acc
```

```python
print(confusion_matrix(y_test,y_pred_bag))
```

```python
print(classification_report(y_test,y_pred_bag))
```

```python
model_result["Bagging Classifier"]=accuracy_score(y_test,y_pred_bag)
```

```python
model_result
```

```python
models=pd.DataFrame({
    "Model" : ["Logistic Default","Logistic with l1","DecisionTreeClassifier","KNN","BaggingClassifier"],
    "Scores" : [log_acc,log_l1_acc,dct_acc,knn_acc,bag_acc]
})
```

```python
models
```

```python
colors=["lightgreen","hotpink","khaki","tomato","skyblue"]
plt.figure(figsize=(16,8))
sns.barplot(x=models["Model"],y=models["Scores"],palette=colors)
plt.show()
```

```python
#from here we get to know that the accuracy of the Bagging Classifier is the Best among all
```
